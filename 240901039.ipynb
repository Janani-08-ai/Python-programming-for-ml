```python
#create an array
import numpy as np
a=np.array([[1,2,4],[5,8,7]])
print("Array created :\n",a)
```

```python
#creating a 3*4 array with all zeros
b=np.zeros((3,4))
print("Array with nall zeros :\n",b)
```

```python
#create a constant value array
c=np.full((3,3),6)
print("Array with all 6s:\n",c)
```

```python
#create an array with random value
d=np.random.random((2,2))
print("Random array:\n",d)
```

```python
#creating a sequence of integers from 0 to 30 with step of 5
import numpy as np
e=np.arange(0,30,5)
print(e)
```

```python
#Reshaping 3*4 array to 2*2*3 array
import numpy as np
arr=np.array([[1,2,3,4],[5,2,4,2],[1,2,0,1]])
newarr=arr.reshape(4,3)
print("\n Original array:\n",arr)
print("Reshaped array[4,3]:\n",newarr)


```

```python
#flatten array
flarr=arr.flatten()
print("\n Original array:\n",arr)
print("\n flattened array:\n",flarr)
```
